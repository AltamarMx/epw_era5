```python
from iertools.read import read_epw
import matplotlib.pyplot as plt
import pandas as pd
from windrose import WindroseAxes
from dateutil.parser import parse
import xarray as xr
```

```python

f = '../epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2007-2021.epw'
epw = read_epw(f,alias=True,warns=False)

f = '../data/memo_point_file.nc'
era = xr.open_dataset(f, engine='netcdf4')
```

# Temperatura ambiente To

```python

fig, ax = plt.subplots(figsize=(12,3))

f1 = parse('2007-05-01')
f2 = f1 + pd.Timedelta('31d')

ax.plot( epw.To,label='07_21',alpha=0.7,)

ax.legend()
ax.set_xlim(f1,f2)
```

# Radiaci'on global Ig, directa Ib y difusa Id

```python

fig, ax = plt.subplots(figsize=(12,3))


ax.plot( epw.Ig,label='Ig')
ax.plot( epw.Ib,label='Ib')
ax.plot( epw.Id,label='Id')

ax.legend()
ax.set_xlim(f1,f2)
```

# Rosa de vientos con velocidad ws, y direcci'on wd

```python
ax = WindroseAxes.from_ax()
ax.bar(epw.Wd, epw.Ws, normed=True, opening=0.8, edgecolor='white',bins=[0., 0.1,1.34,2.67,4])
ax.set_legend()
```
