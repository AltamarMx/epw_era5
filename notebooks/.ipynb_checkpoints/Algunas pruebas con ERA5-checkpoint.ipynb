```python
for i in range(0,int(len(ds.time.values))):
    # Get the terrain height
    slp = ds.t2m[i]-273.15
#     tlp = ds.t2m[i][1]
    # Get the cartopy object and the lat,lon coords
    lons = ds.variables['longitude']
    lats = ds.variables['latitude']

    u = ds.variables['u10'][i]
    v = ds.variables['v10'][i]
    # Create a figure and get the GetAxes object
    # # # # # Make the contour lines and fill them.
    plt.contour(np.asarray(lons), np.asarray(lats), 
               np.asarray(slp), colors="black")               
    plt.contourf(np.asarray(lons), np.asarray(lats), 
                np.asarray(slp),                
                cmap=get_cmap("jet"))
    plt.quiver(np.asarray(lons),np.asarray(lats), np.asarray(u), np.asarray(v))
    # clev850 = np.arange(6000, 10000, 30)
    plt.contour(np.asarray(lons), np.asarray(lats), np.asarray(slp), colors='k',linewidths=2.0, linestyles='solid')           
    plt.scatter(-99.26200,18.83500,c='black', s=100)
    # # # # Add a color bar. The shrink often needs to be set 
    # # # # by trial and error.
    ts = pd.to_datetime(str(ds.variables['time'][i].values))
    d = ts.strftime('"%d-%m-%Y %H:%M:%S"')
    print(d)

    plt.title("ERA5_temperature" + '  ' + d)
    plt.savefig("" + str(i) + ".png")    
    plt.show()
```

```python
for i in range(0,int(len(ds.time.values))):
    # Get the terrain height
    slp = ds.t2m[i]-273.15
#     tlp = ds.t2m[i][1]
    # Get the cartopy object and the lat,lon coords
    lons = ds.variables['longitude']
    lats = ds.variables['latitude']

    u = ds.variables['u10'][i]
    v = ds.variables['v10'][i]
    # Create a figure and get the GetAxes object

    fig = plt.figure(figsize=(10, 10))
    geo_axes = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))

    # #https://ncar.github.io/CAM-chem/examples/maps/plot_projection_conus.html
    # # set up line width
    width = 1.0

    geo_axes.coastlines(resolution='10m', linewidth=width)
    geo_axes.add_feature(cfeature.BORDERS.with_scale("10m"), linewidth=width)
    geo_axes.add_feature(cfeature.STATES.with_scale("10m"), linewidth=width/2)

    # # # Download and add the states and coastlines
    # # # See the cartopy documentation for more on this.
    shapeFilename = "./10m_cultural/10m_cultural/ne_10m_admin_1_states_provinces.shp"
    # # print("Reading states provinces from {}\n".format(shapeFilename))
    reader = cartopy.io.shapereader.Reader(shapeFilename)
    statesProvinces = reader.records()

    mexicoStates = []
    for stateProvince in statesProvinces:
        if (stateProvince.attributes["admin"] == "Mexico"):
            #         print("State province {}".format(stateProvince.attributes["name"]))
            mexicoStates.append(stateProvince.geometry)
    geo_axes.add_geometries(mexicoStates, ccrs.PlateCarree(),edgecolor="black", facecolor='none', linewidth=width/2)

    states = NaturalEarthFeature(category='cultural', 
                             scale='50m', 
                             facecolor='none',
                             name=shapeFilename)
    geo_axes.coastlines('50m', linewidth=0.8)

    # # # # Set the contour levels so that all plots match
    # levels = np.arange(980.,1030.,2.5)

    # # # # # Make the contour lines and fill them.
    plt.contour(np.asarray(lons), np.asarray(lats), 
               np.asarray(slp), colors="black",
               transform=crs.PlateCarree())
    plt.contourf(np.asarray(lons), np.asarray(lats), 
                np.asarray(slp),
                transform=crs.PlateCarree(),
                cmap=get_cmap("jet"))
    plt.colorbar(ax=geo_axes, shrink=.9, location='bottom')
    plt.quiver(np.asarray(lons[::4]),np.asarray(lats[::4]), np.asarray(u[::4,::4]), np.asarray(v[::4,::4]), transform=ccrs.PlateCarree())
    # clev850 = np.arange(6000, 10000, 30)
    plt.contour(np.asarray(lons), np.asarray(lats), np.asarray(slp), colors='k',linewidths=2.0, linestyles='solid', transform=ccrs.PlateCarree())           
    # # # # Add a color bar. The shrink often needs to be set 
    # # # # by trial and error.
    ts = pd.to_datetime(str(ds.variables['time'][i].values))
    d = ts.strftime('"%d-%m-%Y %H:%M:%S"')
    print(d)

    plt.title("ERA5_temperature" + '  ' + d)
#     plt.savefig("" + str(i) + ".jpg")
    
    plt.show()
```

```python
list_date=[20201025,20201026,20201027,20201028,20201029]
for list_date in list_date:
    for i in range(0, 2300, 100):
        # read temperature and scale it to C
        t = mv.read(data=ds, date=list_date,param="t", time=i)
        t = t - 273.16
        # read wind components and compute speed
        u = mv.read(data=ds, date=list_date,param="u", time=i)
        v = mv.read(data=ds, date=list_date, param="v", time=i)
        z = mv.read(data=ds, date=list_date, param="z", time=i)
        r = mv.read(data=ds, date=list_date, param="r", time=i)
        sp = mv.sqrt(u * u + v * v)
        
#         ept = mv.eqpott_p(temperature=t, humidity=r)
        
        # read log of surface pressure
        lnsp = mv.read(data=ds, param="lnsp")    
        # compute temperature advection in K/s units
        t_g = mv.gradient(t)
        adv = -(t_g[0::2]*u+ t_g[1::2]*u)
        # scale advection to 1E-4 K/s units
        adv = adv * 1e4
        
        # get fields on 500 hPa
        level = 500
        vo = mv.read(data=ds,date=list_date,time=i, param="vo", levelist=level)
        zo = mv.read(data=ds,date=list_date,time=i,param="z", levelist=level)
        
        u = mv.read(data=ds, date=list_date,time=i,param="u", levelist=level)
        v = mv.read(data=ds,date=list_date,time=i, param="v", levelist=level)
        # compute the eddy kinetic energy (per unit mass) in m2/s2 units
        u_p = u - mv.mean_ew(u)
        v_p = v - mv.mean_ew(v)
        eke = (u_p**2 + v_p**2) / 2

        # compute absolute vorticity from relative vorticity
        absv = mv.absolute_vorticity(vo)

        
        # define cross section line
        line = [32.465015, -116.473977,32.510575, -115.506570]
        # define vertical axis
        vertical_axis = mv.maxis(
        axis_orientation="vertical",
        axis_type="position_list",
        axis_tick_position_list=[1000, 925, 850, 700, 500, 400, 300, 250, 200],
        axis_tick_label_height=0.4,)

        # define pressure range for cross section
        top_pressure = 200
        bottom_pressure = 1000

        # create cross section for advection
        xs_adv = mv.mcross_sect(
        data=adv, line=line, bottom_level=bottom_pressure, top_level=top_pressure)
        
        
#         # create cross section for ept
#         xs_ept = mv.mcross_sect(
#         data=ept, line=line, bottom_level=bottom_pressure, top_level=top_pressure
#         )

        # define cross section view
        xs_view = mv.mxsectview(
        line=line,
        bottom_level=bottom_pressure,
        top_level=top_pressure,
        vertical_axis=vertical_axis,)

        # define contouring for advection
        cont_adv = mv.mcont(
        legend="on",
        contour="off",
        contour_level_selection_type="interval",
        contour_max_level=35,
        contour_min_level=-35,
        contour_interval=3,
        contour_label="off",
        contour_shade="on",
        contour_shade_colour_method="palette",
        contour_shade_method="area_fill",
        contour_shade_palette_name="colorbrewer_RdBu_16",)
        
        
#         # define contouring for ept
#         cont_ept = mv.mcont(
#         contour_line_colour="charcoal",
#         contour_highlight_colour="charcoal",
#         contour_highlight_thickness=2,
#         contour_level_selection_type="interval",
#         contour_interval=2,
#         contour_label_height=0.4,)

        # define legend
        legend = mv.mlegend(legend_text_font_size=0.4)

        # define title
        vdate = mv.valid_date(t[0])
        title = mv.mtext(
        text_lines=[
            "ERA5 t advection [1E-4 K/s] and ept [K] {}".format(
                vdate.strftime("%Y-%m-%d %H UTC")
            ),
            "",
        ],
        text_font_size=0.5,
        )


        # define the output plot file
        mv.setoutput(mv.png_output(output_name="cross_section_pl_tadv_" +  str(list_date) + str(i)))

        # generate plot
        mv.plot(xs_view, adv, cont_adv, legend, title)
        
        #         graficando temperatura
        # set up the view to plot the data into
        cross_section_view = mv.mxsectview(
        bottom_level=1000.0,
        top_level=1,
        line=line,
        )

        # set up the contouring style
        shading = mv.mcont(
            legend="on",
            contour="off",
            contour_level_count=12,
            contour_label="off",
            contour_shade="on",
            contour_shade_method="area_fill",
            contour_shade_max_level_colour="RGB(0.72,0.059,0.059)",
            contour_shade_min_level_colour="RGB(0.99,0.98,0.98)",
        )

        # set up the title, just to make the font bigger
        title = mv.mtext(text_font_size=0.5)

        # define the output plot file
        mv.setoutput(mv.png_output(output_name="cross_section_pl_data_"+ str(list_date) + str(i)))

        # plot the data into the Cross Section view with visdefs for styling
        mv.plot(cross_section_view, t, shading, title)
        
        #         graficando Vorticidad absoluta


        # define contouring for absolute vorticity
        cont_vo = mv.mcont(
            contour_automatic_setting="style_name",
            contour_style_name="sh_viobrn_fM50t50lst_less",
            legend="on",
            grib_scaling_of_derived_fields="on",
        )

        # define contouring for geopotential
        cont_z = mv.mcont(
            contour_line_thickness=2,
            contour_line_colour="charcoal",
            contour_highlight_colour="charcoal",
            contour_highlight_thickness=4,
            contour_level_selection_type="interval",
            contour_interval=5,
            contour_label_height=0.4,
            grib_scaling_of_derived_fields="on",
        )
          
        # define coastlines
        coast = mv.mcoast(map_coastline_colour="chestnut", map_coastline_thickness=3)
        # define the geographical view
        view = mv.geoview(
            map_projection="polar_stereographic",
            map_area_definition="centre",
            map_vertical_longitude=-127.808472,
            map_centre_latitude=32.256875,
            map_centre_longitude=-115.903152,
            map_scale=30.0e6,
            coastlines=coast,
        )

        # define title
        vdate = mv.valid_date(zo)
        title = mv.mtext(
            text_lines=[
                "ERA5 z [dam] and absv [1E5 1/s] {} hPa {}".format(
                    level, vdate.strftime("%Y-%m-%d %H UTC")
                ),
                "",
            ],
            text_font_size=0.5,
        )

        # define legend
        legend = mv.mlegend(legend_text_font_size=0.4)

        # define output
        mv.setoutput(mv.png_output(output_name="absolute_vorticity_" +  str(list_date) + str(i)))

        # generate plot
        mv.plot(view, absv, cont_vo, zo, cont_z, title, legend)
        
                # define contouring
        cont_eke = mv.mcont(
            legend="on",
            contour="off",
            contour_level_selection_type="level_list",
            contour_level_list=[100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1200, 1500],
            contour_label="off",
            contour_shade="on",
            contour_shade_colour_method="palette",
            contour_shade_method="area_fill",
            contour_shade_palette_name="m_purple2_11",
        )

        cont_z = mv.mcont(
            contour_line_thickness=2,
            contour_line_colour="charcoal",
            contour_highlight_colour="charcoal",
            contour_highlight_thickness=4,
            contour_level_selection_type="interval",
            contour_interval=10,
            contour_label_height=0.4,
            grib_scaling_of_derived_fields="on",
        )

        # define coastlines
        coast = mv.mcoast(
            map_coastline_land_shade="on",
            map_coastline_land_shade_colour="grey",
            map_grid_colour="RGB(0.3843,0.3843,0.3843)",
        )


        # define the geographical view
        view = mv.geoview(
            map_projection="polar_stereographic",
            map_area_definition="centre",
            map_vertical_longitude=-127.808472,
            map_centre_latitude=32.256875,
            map_centre_longitude=-115.903152,
            map_scale=35.0e6,
            coastlines=coast,
        )


        # define title
        vdate = mv.valid_date(zo)
        title = mv.mtext(
            text_lines=[
                "ERA5 z [dam] and EKE [m2/s2] {} hPa {}".format(
                    level, vdate.strftime("%Y-%m-%d %H UTC")
                ),
                "",
            ],
            text_font_size=0.5,
        )

        # define legend
        legend = mv.mlegend(legend_text_font_size=0.4)

        # define output
        mv.setoutput(mv.pdf_output(output_name="eddy_kinetic_energy_"+ str(list_date) + str(i)))

        # generate plot
        mv.plot(view, eke, cont_eke, zo, cont_z, title, legend)
```
