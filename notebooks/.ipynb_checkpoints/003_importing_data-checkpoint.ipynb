```python
ds = xr.open_dataset('../data/cuernavaca.nc', engine='netcdf4')
```

```python
ds
```

```python
t2m_mean = ds['t2m'].mean(dim=['latitude', 'longitude'])

# Graficar t2m promedio en función del tiempo
(t2m_mean-273).plot()
```

```python
ssrd = ds['ssrd'].mean(dim=['latitude', 'longitude'])

# Graficar t2m promedio en función del tiempo

fig, ax = plt.subplots(figsize=(12,4))
ax.plot(ssrd/3600,label='era' ) 
ax.plot(epw.Ig,label='epw')

ax.legend()
```
