```python
from iertools.read import read_epw
import matplotlib.pyplot as plt
import pandas as pd
from windrose import WindroseAxes
```

```python
f = "../epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2004-2018.epw"
tmy_04_18 = read_epw(f,alias=True,warns=False)

f = '../epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2007-2021.epw'
tmy_07_21 = read_epw(f,alias=True,warns=False)

f = '../epw/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.epw'
tmy = read_epw(f,alias=True,warns=False)
```

```python
fig, ax = plt.subplots(figsize=(12,3))

# ax.plot(tmy_04_18.To,label='04_18',alpha=0.7)
ax.plot(tmy_07_21.To,label='07_21',alpha=0.7)
ax.plot(tmy.To,label='tmy',alpha=0.7)

ax.legend()
```
